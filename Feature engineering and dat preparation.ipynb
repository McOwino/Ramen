```python
# import libraries
import pandas as pd
```

```python
# read the .csv from part II as dataframe
df = pd.read_csv('thebiglist_clean_extended.csv')
df
```

```python
# get value counts for brands
top_thirty_brand = df['Brand'].value_counts()[:30]
top_thirty_brand
```

```python
# create another column called the newBrand
new_brand = []

for brand in df['Brand']:
    if brand in top_thirty_brand:
        new_brand.append(brand)
    else:
        new_brand.append('other')
        
df['newBrand']=new_brand  
df
```

```python
len(df['newBrand'].value_counts())
```

```python
len(df['Brand'].value_counts())
```

```python
top_four_style = df['Style'].value_counts()[:4]
top_four_style
```

```python
# loop through style and create a newstyle column
new_style = []

for style in df['Style']:
    if style in top_four_style:
        new_style.append(style)
    else:
        new_style.append ('other')
        
df['newStyle']=new_style
df
```

```python
df['newStyle'].value_counts()
```

```python
#examine and choose the top countries 
df['Country'].value_counts()[:9].sum()
```

```python
top_ten_countries = df['Country'].value_counts()[:9]
top_ten_countries
```

```python
new_counry = []

for country in df['Country']:
    if country in top_ten_countries:
        new_counry.append(country)
    else:
        new_counry.append('other')
        
df['newCountry']=new_counry  
df
```

```python
df['newCountry'].value_counts()
```

```python
brand_dummy = pd.get_dummies(df['newBrand'], drop_first=True, 
                          prefix="from")
brand_dummy
```

```python
style_dummy = pd.get_dummies(df['newStyle'], drop_first=True,
                            prefix="is")
style_dummy
```

```python
country_dummy = pd.get_dummies(df['newCountry'], drop_first=True,
                              prefix="in")
country_dummy
```

```python
df_temp=df.drop(['Brand','Country','Style','Variety','newCounry','newBrand','newCountry','newStyle'], axis= 1)
df_temp
```

```python
df_final = pd.concat([df_temp, brand_dummy, style_dummy, country_dummy], axis= 1) 
df_final
```

```python
df_final.to_csv('thebiglist_final.csv')
pd.read_csv('thebiglist_final.csv')
```

### Summary
1. Changed low_frequency values into "others".
2. Dummified the new columns.
3. Droppped unnecessary columns.
4. Joined everything into a final DataFrame.
